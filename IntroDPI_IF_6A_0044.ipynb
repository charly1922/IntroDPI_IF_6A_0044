```python
from google.colab import drive
drive.mount('/content/drive')

import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Load gambar
img = cv2.imread('/content/drive/MyDrive/kucing.jpg')

if img is None:
    print("Error: Gambar tidak ditemukan")
else:
    img = cv2.resize(img, (400, 400))

    print("Shape:", img.shape)
```

```python
 # Grayscale
 img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_crop = img_gray[200:208, 180:188]
print("=== MATRIKS f(x,y) ===")
print(img_crop)
print("Contoh:", f"f(180,200) = {img_gray[200,180]}")
```

```python
# Visualisasi
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img_gray, cmap='gray', vmin=0, vmax=255)
axes[0].set_title('grayscale')
axes[0].axis('off')

rect = patches.Rectangle((180, 200), 8, 8, linewidth=2, edgecolor='red', facecolor='none')
axes[0].add_patch(rect)
axes[0].text(190, 195, 'area crop', color='red', fontsize=10)

axes[1].imshow(img_crop, cmap='gray', vmin=0, vmax=255)

for i in range(8):
        for j in range(8):
            nilai = img_crop[i, j]
            text_color = 'white' if nilai < 70 else 'black'
            axes[1].text(j, i, str(nilai),
                         ha='center', va='center',
                         fontsize=8, color=text_color, fontweight='bold')

axes[1].set_title('8x8 pixel\nf(x,y)')
axes[1].axis('off')

plt.suptitle('Representasi f(x,y)', fontweight='bold')
plt.tight_layout()
plt.show()
```
